```python
from autogen import ConversableAgent
```

```python
from google.colab import userdata
import os
api_key=userdata.get('Gemini_main')
os.environ["GEMINI_API_KEY"] = api_key
```

```python
llm_config = {
  "config_list": [
    {
      "api_type": "gemini",
      "model": "gemini-2.0-flash-lite",
      "api_key": api_key,
      "api_type": "google"
    }
  ],
}
```

```python
agent1 = ConversableAgent(
    name="Saish",
    system_message=
    "Your name is Saish and you are a very intelligent researcher.You help in teaching concepts concisely ",
    llm_config=llm_config,
    human_input_mode="NEVER",
)

agent2 = ConversableAgent(
    name="Cristiano",
    system_message=
    "Your name is Cristiano and you are sarcastic student trying to learn but not interested in studies as you are a professional footballer",
    llm_config=llm_config,
    human_input_mode="NEVER",
)
```

```python
chat_result = agent2.initiate_chat(
    recipient=agent1,
    message="What is Ai?",
    max_turns=4,
)
```

## Chat History

```python
import pprint

pprint.pprint(chat_result.chat_history)
```

## Chat Cost

```python
pprint.pprint(chat_result.cost)
```

## Chat Summary(Not a great one)

```python
pprint.pprint(chat_result.summary)
```

## For better chat summary

```python
chat_result = agent2.initiate_chat(
    recipient=agent1,
    message="What is Ai?",
    max_turns=4,
    summary_method="reflection_with_llm",
    summary_prompt="Summarize the conversation",
)
```

```python
pprint.pprint(chat_result.summary)
```

##Chat Termination

```python
agent1 = ConversableAgent(
    name="Saish",
    system_message=
    "Your name is Saish and you are a very intelligent researcher.You help in teaching concepts concisely "
     "When you're ready to end the conversation, say 'Goodbye'.",
    llm_config=llm_config,
    human_input_mode="NEVER",
    is_termination_msg=lambda msg: "Goodbye" in msg["content"]
)

agent2 = ConversableAgent(
    name="Cristiano",
    system_message=
    "Your name is Cristiano and you are sarcastic student trying to learn but not interested in studies as you are a professional footballer"
    "When you're ready to end the conversation, say 'Goodbye'.",
    llm_config=llm_config,
    human_input_mode="NEVER",
    is_termination_msg=lambda msg: "Goodbye" in msg["content"] ## the termination statement should be same and if not then it will go in a endless loop of convo
)
```

```python
chat_result = agent2.initiate_chat(
    recipient=agent1,
    message="What is Ai?"
)
```
